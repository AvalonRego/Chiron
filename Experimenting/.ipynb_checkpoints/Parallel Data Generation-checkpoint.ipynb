```python
from joblib import Parallel, delayed
import numpy as np
import time

import warnings

# Suppress SyntaxWarning specifically from jax
warnings.filterwarnings(
    "ignore",
    message="Jitted function has invalid argnames",
    category=SyntaxWarning
)


def process_sample(i, detector):

    samples = 10  # Number of samples for this configuration
    photon_propagator_configuration = MockPhotonPropagatorConfiguration(
        resolution=18000,
        medium=MediumEstimationVariant.PONE_OPTIMISTIC,
        max_memory_usage=int(2147483648 / 8)
    )

    configuration = DatasetConfiguration(
        detector=Hex_Configuration,
        generators=[
            GenerationConfiguration(
                generator=EventGeneratorConfiguration(
                    type=EventType.CASCADE,
                    spectrum=UniformSpectrumConfiguration(
                        log_minimal_energy=2.0,
                        log_maximal_energy=5.5
                    ),
                    source_propagator=photon_propagator_configuration
                ),
                number_of_samples=samples
            )
        ],
        storage=HDF5StorageConfiguration(
            data_path=f'data/Parallel_testC/{i}.h5',
            read_only=False
        )
    )

    start_time = time.time()
    print(f"Starting sample generation for configuration {i} with {samples} samples.")
    collection = generate(configuration)  # Run the generate process
    end_time = time.time() - start_time

    print(f"Finished processing configuration {i}. Time taken: {end_time:.2f}s")
    return end_time, samples


def main():  # Replace with actual import for Hex_Configuration

    detector = Hex_Configuration  # Assign detector configuration
    num_workers = 2  # Number of parallel workers
    iterations = range(1, 3)  # Define the range of configurations

    # Use Parallel processing
    results = Parallel(n_jobs=num_workers, verbose=10)(
        delayed(process_sample)(i, detector) for i in iterations
    )

    # Separate times and samples into individual arrays
    times, samples = zip(*results)
    times = np.array(times)
    samples = np.array(samples)

    print("Processing complete.")
    print("Times:", times)
    print("Samples:", samples)
    return times, samples


if __name__ == "__main__":
    times, samples = main()
```
